```python
pm10fil=pm10fil.rename(columns={"codigoSerial":"Codigo"})
print(pm10fil)
```

```python
conteo_puntospm10 = superado_pm10['geometry'].value_counts().reset_index()
conteo_puntospm10.columns = ['geometry', 'conteo']
print(conteo_puntospm10)
```

```python
#volviendo conteo puntos geodataframe
conteo_puntospm10=gpd.GeoDataFrame(conteo_puntospm10, geometry="geometry", crs="EPSG:4326")
```

```python
#uniendo conteo puntos con municipios
puntos_en_municipiospm10 = gpd.sjoin(conteo_puntospm10, municipios, how='inner',predicate='within')

# Calcular el conteo total por municipio
conteo_por_municipiopm10 = puntos_en_municipiospm10.groupby('MPIO_NOMBR').size().reset_index(name='conteos')

# Unir este conteo con el GeoDataFrame original de municipios
municipios_con_conteopm10 = municipios.merge(conteo_por_municipiopm10, on='MPIO_NOMBR', how='left')

# Rellenar NaN con 0 para municipios sin puntos
municipios_con_conteopm10['conteos'] = municipios_con_conteopm10['conteos'].fillna(0)
print(municipios_con_conteopm10)
```

```python
municipios_con_conteopm10 = municipios_con_conteopm10.to_crs(epsg=3857)

fig, ax = plt.subplots(1, 1, figsize=(12, 10))

cmap = 'Reds'
vmin = municipios_con_conteopm10['conteos'].min()
vmax = municipios_con_conteopm10['conteos'].max()
norm = mpl.colors.Normalize(vmin=vmin, vmax=vmax)

municipios_con_conteopm10.plot(
    column='conteos',
    ax=ax,
    cmap=cmap,
    edgecolor='gray',
    linewidth=0.3,
    missing_kwds={'color': 'lightgrey'}
)

ctx.add_basemap(ax, source=ctx.providers.CartoDB.Positron)

sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)
sm._A = []
cbar = fig.colorbar(sm, ax=ax, orientation='horizontal', shrink=0.5, pad=0.02)

# Mostrar solo valores enteros en la barra
cbar.locator = MaxNLocator(integer=True)
cbar.update_ticks()

cbar.set_label("Número de estaciones PM10 que sobrepasan ")

for idx, row in municipios_con_conteopm10.iterrows():
    if row['geometry'].is_empty:
        continue
    x, y = row['geometry'].centroid.coords[0]
    ax.text(x, y, row['MPIO_NOMBR'], fontsize=6, ha='center', va='center', color='black')

ax.set_title('Conteo de las estaciones que exceden PM 10 por municipio', fontsize=16)
ax.set_axis_off()
plt.tight_layout()
plt.show()
```

```python
#Densidad 
# Asegurarte de estar en un sistema proyectado (como metros)
municipios_con_conteopm10 = municipios_con_conteopm10.to_crs(epsg=3857)

# Calcular área en km²
municipios_con_conteopm10["area_km2"] = municipios_con_conteopm10.geometry.area / 1e6

# Mostrar un ejemplo
print(municipios_con_conteopm10[["MPIO_NOMBR", "area_km2"]].head())
```

```python
municipios_con_conteopm10["densidad"]=municipios_con_conteopm10["conteos"]/municipios_con_conteopm10["area_km2"]
```

```python
import pandas as pd
import statsmodels.formula.api as smf

# Asegúrate de que las columnas estén bien tipadas
df_final['superado'] = df_final['superado'].astype(int)
df_final['Codigo_calidad'] = df_final['Codigo_calidad'].astype('category')

# Diccionario para guardar los modelos por estación
modelos_por_estacion = {}

# Recorrer cada estación y ajustar el modelo logístico
for cod in df_final['Codigo_calidad'].unique():
    datos_estacion = df_final[df_final['Codigo_calidad'] == cod]
    
    # Comprobar si hay variabilidad en la variable dependiente (superado)
    if datos_estacion['superado'].nunique() < 2:
        print(f"⚠️ Estación {cod} omitida: sin variabilidad en 'superado'")
        continue
    
    try:
        modelo = smf.logit('superado ~ temperatura + pp_diaria', data=datos_estacion).fit(disp=0)
        modelos_por_estacion[cod] = modelo
        print(f"\n📍 Estación: {cod}")
        print(modelo.summary())
    except Exception as e:
        print(f"⚠️ Error al ajustar modelo para la estación {cod}: {e}")
```

```python
from statsmodels.api import Logit
from sklearn.preprocessing import StandardScaler

# 1. Estandarizar variables
scaler = StandardScaler()
df_finalpm10[['temperatura', 'pp_diaria']] = scaler.fit_transform(
    df_finalpm10[['temperatura', 'pp_diaria']]
)

# 2. Modelo con parámetros de optimización ajustados
modelos_por_estacion = {}
for cod in df_finalpm10['Codigo_calidad'].unique():
    datos_estacion = df_finalpm10[df_finalpm10['Codigo_calidad'] == cod]
    
    # Verificar balance de clases
    class_counts = datos_estacion['superado'].value_counts()
    if len(class_counts) < 2 or class_counts.min() < 5:
        print(f"⚠️ Estación {cod} omitida: clases desbalanceadas {dict(class_counts)}")
        continue
    
    try:
        modelo = Logit(
            datos_estacion['superado'],
            datos_estacion[['temperatura', 'pp_diaria']],
            missing='drop'
        ).fit(
            method='bfgs',  # Alternativa a Newton-Raphson
            maxiter=1000,   # Aumentar iteraciones
            disp=False
        )
        modelos_por_estacion[cod] = modelo
        print(f"\n✅ Estación {cod}: Convergencia exitosa")
        print(modelo.summary())
    except Exception as e:
        print(f"❌ Error en estación {cod}: {str(e)}")
```
